- One possible soluion: **Quantized models** 

During Inference: 

**Quantization error:** Difference between the orginal and dequantized values. The goal of the quantization is to reduce this error as much as we can

### Paramter Efficient Fine-Tuning (PEFT):

Challenges of Full Fine-Tuning:
- Memory-Intensive Requirements
- Catastrophic Forgetting
- Storage and Computational Costs

**PEFT** seeks to update only a small subset of the model parameters, making the fine-tuning process more manageable in terms of memory requirements and computational resources.
